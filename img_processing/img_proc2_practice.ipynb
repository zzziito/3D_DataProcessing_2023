Geometic Transform

Translation

image.png

```python
trans_x = 10
trans_y = 40

img_trans = img.copy() * 0.0

h, w, ch = img.shape

# set the basis image: original image
for v in range(h):
  for u in range(w):
    target_u = u + trans_x
    target_v = v + trans_y
    if target_u >= 0 and target_u < w and target_v >= 0 and target_v < h:
      img_trans[target_v, target_u, :] = img[v, u, :]


cv2_imshow(img_trans)
```

Scaling

Forward transform



```python
scale = 1.5

img_scale1 = img.copy() * 0.0

h, w, ch = img.shape

# scale center: (0,0)
# set the basis image: original image
for v in range(h):
  for u in range(w):
    target_u = int(u * scale)
    target_v = int(v * scale)
    if target_u >= 0 and target_u < w and target_v >= 0 and target_v < h:
      img_scale1[target_v, target_u, :] = img[v, u, :]

cv2_imshow(img_scale1)
```

Rotation: Forward - Center

```python
deg = 10 * np.pi / 180.0

img_rot_f = img.copy() * 0.0

h, w, ch = img.shape
vv = int(h/2)
uu = int(w/2)

# scale center: (0,0)
# set the basis image: original image
for v in range(h):
  for u in range(w):
    target_u = int((u-uu) * np.cos(deg) - (v-vv) * np.sin(deg) + uu) 
    target_v = int((u-uu) * np.sin(deg) + (v-vv) * np.cos(deg) + vv)
    if target_u >= 0 and target_u < w and target_v >= 0 and target_v < h:
      img_rot_f[target_v, target_u, :] = img[v, u, :]

cv2_imshow(img_rot_f)
```
